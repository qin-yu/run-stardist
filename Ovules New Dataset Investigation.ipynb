```python
import os
import glob
import h5py
import tifffile
from tqdm import tqdm
from pprint import pprint
from natsort import natsorted
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import zoom
```

```python
input_folder_path = "/g/kreshuk/yu/Datasets/TMody2021Ovules/.original/"
raw_file_list = natsorted(glob.glob(input_folder_path + "*_n_stain.tif"))
lab_file_list = natsorted(glob.glob(input_folder_path + "*_n_stain_segmented_corrected.tif"))

pprint(raw_file_list);print()
pprint(lab_file_list)
```

```python
def read_file(raw_file, lab_file):
    raw = tifffile.imread(raw_file)
    lab = tifffile.imread(lab_file)
    lab_unique = len(np.unique(lab))
    print(f'File: {raw_file.split("/")[-1]}')
    print(raw.dtype, raw.shape, lab.dtype, lab.shape, lab_unique)
    assert raw.shape == lab.shape
    
    return raw, lab
```

```python
def show_raw_lab(raw, lab, i):
    fig, axs = plt.subplots(1, 2, figsize=(16, 8))
    ax = axs[0]
    ax.imshow(raw[i])
    ax = axs[1]
    ax.imshow(lab[i])
```

```python
def show_flipped(raw, lab, i, axis=1):
    fig, axs = plt.subplots(1, 2, figsize=(16, 8))
    
    ax = axs[0]
    ax.imshow(raw[i])
    ax.imshow(np.flip(lab, axis=axis)[i], alpha=0.5)
    
    ax = axs[1]
    i += 10
    ax.imshow(raw[i])
    ax.imshow(np.flip(lab, axis=axis)[i], alpha=0.5)
    i -= 10
```
